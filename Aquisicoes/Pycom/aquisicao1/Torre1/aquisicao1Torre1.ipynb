#### Filtrando os dados:

A matriz do End Node veio muito maior que a do Gateway. Como se sabe, o frame count **n** do End Node corresponde ao frame count __n+1__ do Gateway. Então o código abaixo percorre a matriz do End Node e remove tudo que for texto ("Sem resposta" ou "inicio") e tudo que não for um frame count igual a um frame count +1 do gateway. Ainda pode ser que um determinado frame count +1 não exista na matriz do End Node, talvez por perda de pacote. Então alguns dados da matriz do gateway também deverão ser excluidos.

```python
# Criando a matriz final do end node e os indexes que deverão ser excluidos da matriz do gateway
matrizEndNode = []
indexesExcluir = []
perda = 0

# Percorrendo a matriz do gateway
for index, vGateway in enumerate(matrizGateway):
    # Guardando o frame count atual e setando a variavel que vai dizer 
    # se existe o frame count +1 na matriz do endnode
    fCntAtual = vGateway[0]
    existe = 0
    
    # Percorrendo a matriz do endnode
    for vEndNode in matrizENPreliminar:
        # Se tiver um frame count +1 e o dado for um inteiro, então existe
        if vEndNode[0] == str(int(fCntAtual) + 1):
            try:
                inteiro = int(vEndNode[1])
                matrizEndNode.append(vEndNode)
                existe = 1
                break
            except:
                pass         
            
    # Se no final não existir, então é um index do gateway que deverá ser excluido
    if existe == 0:
        indexesExcluir.append(index)

# Removendo os indexes do gateway que não tem um correspondente frame count +1 no endnode
for i in reversed(indexesExcluir):
    del matrizGateway[i]
    perda += 1
        
# Removendo os ultimos dados do gateway até o tamanho das duas matrizes ser igual
while len(matrizEndNode) != len(matrizGateway):
    matrizGateway = matrizGateway[:-1]
```
