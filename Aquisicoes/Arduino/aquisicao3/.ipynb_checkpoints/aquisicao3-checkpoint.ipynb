## Importando os dados dos CSV's

O ".T" ao final irá retornar a matriz transposta. Dessa forma, vai ser mais fácil de plotar. Além disso, o "[1:]" está ignorando a primeira linha, já que é o cabeçalho.

```python
endNodeT2A3       = np.genfromtxt('torre2Andar3/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeDCE        = np.genfromtxt('dce/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeBlocoBA5   = np.genfromtxt('blocoBAndar5/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeBiblioteca = np.genfromtxt('biblioteca/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeBlocoLA4   = np.genfromtxt('blocoLAndar4/CSV/ENDNODE.csv', delimiter=',')[1:].T

gatewayT2A3       = np.genfromtxt('torre2Andar3/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayDCE        = np.genfromtxt('dce/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayBlocoBA5   = np.genfromtxt('blocoBAndar5/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayBiblioteca = np.genfromtxt('biblioteca/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayBlocoLA4   = np.genfromtxt('blocoLAndar4/CSV/GATEWAY.csv', delimiter=',')[1:].T
```

##### Index endNode

endNode[0] -> fCnt

endNode[1] -> data

endNode[2] -> adr

endNode[3] -> dr

endNode[4] -> frequency

##### Index gateway

gateway[0] -> frequency

gateway[1] -> bandwidth

gateway[2] -> spreadingFactor

gateway[3] -> codeRate

gateway[4] -> timestamp

gateway[5] -> rssi

gateway[6] -> loraSnr

gateway[7] -> channel

gateway[8] -> rfChain

gateway[9] -> latitude

gateway[10] -> longitude

gateway[11] -> altitude

gateway[12] -> adr

gateway[13] -> adrAckReq

gateway[14] -> fCnt

gateway[15] -> fOpts

## Perda de Pacote

Packet Delivery Ratio: taxa de entrega de pacotes -> recebidos/enviados

### T2A3

```python
pdrT2A3 = (len(endNodeT2A3[0]) - 1)/endNodeT2A3[0][-1]

print("Enviados:", int(endNodeT2A3[0][-1]))
print("Recebidos:", len(endNodeT2A3[0]) - 1)
print("\nPDR:", pdrT2A3)
```

### Data Rate

```python
# FUNÇÃO PARA PLOTAR CADA UMA DAS BARRAS
def plotar(vetorDR, vetorRSSI, cont):
    # COMEÇO ALMENTANDO 1 NO CONTADOR PARA SEPARAR AS BARRAS DE LUGARES DIFERENTES
    cont += 1
    for dr in np.unique(vetorDR):
        # A COR DEPENDE DO DR
        if dr == 0:
            cor = "#c2d1f0"
        elif dr == 1:
            cor = "#85a3e9"
        elif dr == 2:
            cor = "#4775d1"
        elif dr == 3:
            cor = "#2952a3"
        elif dr == 4:
            cor = "#193366"
        else:
            cor = "#0f1f3d"
        
        # PLOTANDO A BARRA
        ax1.bar(cont, dr + 1, color=cor, width=1.01)
        
        # AGORA É FAZER A INCERTEZA DO RSSI
        vetorRSSIAtual = [vetorRSSI[i] for i, dataRate in enumerate(vetorDR) if dataRate == dr]
        ax2.errorbar(cont, np.mean(vetorRSSIAtual), yerr=2*np.std(vetorRSSIAtual), color="#ff0066")
        ax2.scatter(cont, np.mean(vetorRSSIAtual), color="#ff0066")
    
        # INCREMENTANDO O CONTADOR E RETORNANDO-O
        cont +=1
    return cont

# CRIANDO OS DOIS EIXOS
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

# PLOTANDO
cont = 0
cont = plotar(endNodeT2A3[3], gatewayT2A3[5], cont)
cont = plotar(endNodeDCE[3], gatewayDCE[5], cont)
cont = plotar(endNodeBiblioteca[3], gatewayBiblioteca[5], cont)
cont = plotar(endNodeBlocoBA5[3], gatewayBlocoBA5[5], cont)
cont = plotar(endNodeBlocoLA4[3], gatewayBlocoLA4[5], cont)

# PARÂMETROS
plt.title("Data Rate e RSSI")
ax1.set_ylabel("Data Rate")
ax2.set_ylabel("RSSI Médio")
ax1.set_xlabel("Posição")
ax1.grid(True, axis='y', alpha=0.5)

# ARRUMANDO O EIXO Y
ax1.set_ylim(0.5,6.5)
ax1.set_yticklabels(["-1","0","1","2","3","4","5"])
ax2.set_ylim(-112, -88)
ax2.set_yticks(np.arange(-110,-89,4))

# ARRUMANDO O EIXO X
plt.xticks([2, 5.5, 9, 12, 14],["Posição 1", "Posição 2", "Posição 3", "Posição 4", "Posição 5"])

# SALVANDO
plt.savefig('graficoDR.pdf', format='pdf', bbox_inches = 'tight')

# EXIBINDO
plt.plot()
```
