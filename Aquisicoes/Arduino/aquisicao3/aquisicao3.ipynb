## Importando os dados dos CSV's

O ".T" ao final irá retornar a matriz transposta. Dessa forma, vai ser mais fácil de plotar. Além disso, o "[1:]" está ignorando a primeira linha, já que é o cabeçalho.

```python
endNodeT2A3       = np.genfromtxt('torre2Andar3/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeDCE        = np.genfromtxt('dce/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeBlocoBA5   = np.genfromtxt('blocoBAndar5/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeBiblioteca = np.genfromtxt('biblioteca/CSV/ENDNODE.csv', delimiter=',')[1:].T
endNodeBlocoLA4   = np.genfromtxt('blocoLAndar4/CSV/ENDNODE.csv', delimiter=',')[1:].T

gatewayT2A3       = np.genfromtxt('torre2Andar3/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayDCE        = np.genfromtxt('dce/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayBlocoBA5   = np.genfromtxt('blocoBAndar5/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayBiblioteca = np.genfromtxt('biblioteca/CSV/GATEWAY.csv', delimiter=',')[1:].T
gatewayBlocoLA4   = np.genfromtxt('blocoLAndar4/CSV/GATEWAY.csv', delimiter=',')[1:].T
```

##### Index endNode

##### Index gateway

gateway[0] -> frequency

gateway[1] -> bandwidth

gateway[2] -> spreadingFactor

gateway[3] -> codeRate

gateway[4] -> timestamp

gateway[5] -> rssi

gateway[6] -> loraSnr

gateway[7] -> channel

gateway[8] -> rfChain

gateway[9] -> latitude

gateway[10] -> longitude

gateway[11] -> altitude

gateway[12] -> adr

gateway[13] -> adrAckReq

gateway[14] -> fCnt

gateway[15] -> fOpts
